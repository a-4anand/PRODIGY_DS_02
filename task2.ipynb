```python
## Detemining the number of items in each unique category
df_items = pd.DataFrame(df['hpi_type'].value_counts())     ## Create a DataFrame to see number of Items by Category
df_items['ItemCategory'] = df_items.index                  ## assign "Category" as index
df_items.columns = ['No. of Houses', 'House Type']

df_items.head()
```

```python
sns.set(rc={'axes.facecolor':'none','axes.grid':False,'xtick.labelsize':12,'ytick.labelsize':12, 'figure.autolayout':True})
plt.figure(figsize=(10,8))
my_pal = ('#c7e9b4', '#40E0D0', '#FAEBD7')

## Countplot to show number of Items in each category segment
plt.subplot(2,2,1)
sns.countplot(x="hpi_type", palette=my_pal, data=df, order=df['hpi_type'].value_counts().index).set(title='hpi type(Count)')

## Pie diagram to show % share of each category segment
plt.subplot(2,2,2)
colors = ('#c7e9b4', '#40E0D0', '#FAEBD7')
df['hpi_type'].value_counts().plot(kind='pie', colors=my_pal,legend=None, ylabel='', autopct='%1.0f%%').set(title='Segments in hpi_tgype variable (in %)')
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['period'], df['index_nsa'], label='Index NSA')
plt.plot(df['period'], df['index_sa'], label='Index SA')
plt.title('HPI Index Over Time')
plt.xlabel('Period')
plt.ylabel('Index Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python

sns.set_style("whitegrid")

# Create figure and axes
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Plot index_nsa over time
sns.lineplot(data=df, x='yr', y='index_nsa', ax=axes[0])
axes[0].set_title('Index NSA Over Time')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Index NSA')

# Plot index_sa over time
sns.lineplot(data=df, x='yr', y='index_sa', ax=axes[1])
axes[1].set_title('Index SA Over Time')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Index SA')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Create figure and axes
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# Plot index_nsa distribution across levels
sns.boxplot(data=df, x='level', y='index_nsa', ax=axes[0])
axes[0].set_title('Index NSA Distribution Across Levels')
axes[0].set_xlabel('Level')
axes[0].set_ylabel('Index NSA')

# Plot index_sa distribution across levels
sns.boxplot(data=df, x='level', y='index_sa', ax=axes[1])
axes[1].set_title('Index SA Distribution Across Levels')
axes[1].set_xlabel('Level')
axes[1].set_ylabel('Index SA')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
